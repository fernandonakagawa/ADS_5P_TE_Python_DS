```python
#6 faça uma função que filtra os carros por cilindro e salva em outro arquivo CSV
def filtro_por_cilindro(df, n_cilindros):
  df_cilindros = df.loc[df['num-of-cylinders'] == n_cilindros]
  df_cilindros.to_csv('cilindros.csv')

filtro_por_cilindro(df, 'eight')
```
